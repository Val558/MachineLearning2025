3. Cual es el genero con mayor popularidad y cual es el genero mas popular

```python
Genero_mayor_pupularidad = data.groupby("genre")["popularity"].mean().sort_values(ascending=False)
```

```python
Genero_mayor_pupularidad.head()
```

```python
data[data["genre"] == data["genre"].max()]
```

4. Cual es el artista que tiene mas canciones con popularidades arriba de 70

```python
data[data["popularity"] > 70].groupby("artist").count().sort_values(by="popularity", ascending=False)
```

```python
data[data["popularity"] > 70].groupby("artist").count().sort_values(by="popularity", ascending=False).head()
```
